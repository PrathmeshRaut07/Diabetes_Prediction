```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
asc=[]
name=[]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
asc.append(accuracy_score(y_test,y_pred))
```

```python
name.append('Logistic Regression')
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier_1=RandomForestClassifier(n_estimators=15,random_state=32)
classifier_1.fit(x_train,y_train)
y_pred_1=classifier_1.predict(x_test)
accuracy_score(y_test,y_pred_1)
```

```python
asc.append(accuracy_score(y_test,y_pred_1))
name.append('RandomForestClassifier')
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier_3=DecisionTreeClassifier(criterion='gini',random_state=32)
classifier_3.fit(x_train,y_train)
y_pred_3=classifier_3.predict(x_test)
accuracy_score(y_test,y_pred_3)
```

```python
asc.append(accuracy_score(y_test,y_pred_3))
name.append('DecisionTreeClassifier')
```

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
asc.append(accuracy_score(y_test,y_pred))
name.append('Naive Bayes Classifier')
```

```python
from sklearn.neighbors import KNeighborsClassifier
classs=KNeighborsClassifier(n_neighbors=10)
classs.fit(x_train,y_train)
y_pred=classs.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
asc.append(accuracy_score(y_test,y_pred))
name.append('KNeighborsClassifier')
```

```python
from sklearn.svm import SVC
classifier_5=SVC(kernel='linear',random_state=0)
classifier_5.fit(x_train,y_train)
y_pred_5=classifier_5.predict(x_test)
accuracy_score(y_test,y_pred_5)
```

```python
asc.append(accuracy_score(y_test,y_pred_5))
name.append('Linear SVC')
```

```python
fd=pd.DataFrame({'Classifiers':name,'Accuracy':asc})
```

# Applying Deep Learning Model for more accuracy test ANN

```python
import tensorflow as tf
```

```python
ann=tf.keras.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=128))
```

```python
ann.add(tf.keras.layers.Dense(units=400,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=20,epochs=100)
```

```python
y_pred=ann.predict(x_test)
y_pred=(y_pred>0.5)
y_pred=y_pred.astype(int)
```

```python
accuracy_score(y_test,y_pred)
```
